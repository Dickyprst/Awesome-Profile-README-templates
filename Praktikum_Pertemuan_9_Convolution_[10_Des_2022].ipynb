```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('dicky1.jpeg')
cv2_imshow(img)
kernel1 = np.array([[0, -1, 0],
 [-1, 5,-1],
 [0, -1, 0]])
filter1 = cv2.filter2D(img,-
1,kernel1) #dalam open cv untuk menggunakan konvolusi menggunakan cv2.filter
cv2_imshow(filter1)
```

```python
# perbesar ukuran hasil plotting jika diperlukan
plt.rcParams["figure.figsize"] = (20,20)

img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #harus konversi dulu dari BGR to RGB kalau mau make plot
img2 = cv2.cvtColor(filter1, cv2.COLOR_BGR2RGB) #harus konversi dulu dari BGR to RGB kalau mau make plot

plt.subplot(121), plt.imshow(img1), plt.title("Citra Original") #plot memiliki 1 baris dan 2 kolom. citra akan diletakkan pada plot pertama
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(img2), plt.title("Hasil Konvolusi") #plot memiliki 1 baris dan 2 kolom. citra akan diletakkan pada plot kedua
plt.xticks([]), plt.yticks([])
```
